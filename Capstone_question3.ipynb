#### Importing Pandas

```python
import pandas as pd
```

#### Reading from URL

```python
url = 'https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M'
df_list = pd.read_html(url)
```

#### Extracting only relevant Data

```python
df=df_list[0]
df
```

#### Cleaning the Dataframe to remove rows with "Not Assigned" in the column Borough

```python
ndf=df[df.Borough != "Not assigned"].groupby(['Postal Code', 'Borough', 'Neighbourhood'])
```

##### Reseting the Index

```python
ndf1=pd.DataFrame(ndf.size().reset_index())
ndf2=ndf1[['Postal Code', 'Borough', 'Neighbourhood']]
ndf2
```

#### Showing the shape of the dataframe

```python
ndf2.shape
```

```python
df_latlong = pd.read_csv("Geospatial_Coordinates.csv")
df_latlong
```

#### Merged the 2 Data Frames

```python
df_merged = ndf2.merge(df_latlong)
df_merged
```

```python
import folium
```

```python
venues_map = folium.Map(location=[43.704608, -79.397153], zoom_start=13) 
# generate map centred around the Conrad Hotel

# add the Italian restaurants as blue circle markers
for lat, lng in zip(df_merged.Latitude, df_merged.Longitude):
    folium.features.CircleMarker(
        [lat, lng],
        radius=5,
        color='blue',
        fill = True,
        fill_color='blue',
        fill_opacity=0.6
    ).add_to(venues_map)


# display map
venues_map
```
